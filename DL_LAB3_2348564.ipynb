 Starting with an initial price of 20, we undergo three rounds of momentum-based optimization using a learning rate (α = 0.1) and momentum (β = 0.9):

In the first iteration:
- We calculate the gradient, which equals (w - 30) resulting in -10.
- Then, we update the velocity to 1.
- The price is updated to 21.

In the second iteration:
- We calculate the gradient as (w - 30), which equals -9.
- Subsequently, we update the velocity to 1.8 and the price to 22.8.

In the third iteration:
- We calculate the gradient, resulting in -7.2.
- The velocity is updated to 2.52, and the price is then updated to 25.32.

After three iterations, the updated price approximates 25.32, and we have the option to continue refining the price further without assuming any specific value. [citation:1]  

```python

def revenue_function(w):
    return 0.5 * (w**2) - 30 * w + 100


def derivative_revenue_function(w):
    return (1 * w) - 30


def SGD_momentum_Optimizer(initial_price, learning_rate, momentum,iterations):
    weight_value = initial_price
    velocity = 0

    for i in range(iterations):
        gradient = derivative_revenue_function(weight_value)
        velocity = momentum * velocity + learning_rate * gradient
        weight_value = weight_value - velocity

        print("Iterations: " + str(i + 1) + " weight_value = " + str(weight_value) + "  Optimial Price: " + str(np.abs(revenue_function(weight_value))))
```

```python
import numpy as np

def revenue_function(w):
    """Calculate the revenue function."""
    return 0.5 * w**2 - 30 * w + 100

def gradient(w):
    """Calculate the gradient of the revenue function."""
    return w - 30

def momentum_optimization(initial_w, learning_rate=0.1, momentum=0.9, num_iterations=3):
    w = initial_w
    velocity = 0

    for i in range(num_iterations):
        grad = gradient(w)
        velocity = momentum * velocity - learning_rate * grad
        w += velocity

        print(f"Iteration {i+1}: w = {w:.4f}, Revenue = {revenue_function(w):.4f}")

    return w

# Initial price
initial_price = 20

# Perform momentum-based optimization
optimal_price = momentum_optimization(initial_price, learning_rate=0.1, momentum=0.9)

print(f"\nOptimal price (w) that minimizes revenue lost: {optimal_price:.4f}")
```

```python
le = LabelEncoder()
data['gender'] = le.fit_transform(data['gender'])
data.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
data[['tot_bilirubin', 'direct_bilirubin', 'tot_proteins', 'albumin', 'ag_ratio', 'sgpt', 'sgot', 'alkphos']] = scaler.fit_transform(data[['tot_bilirubin', 'direct_bilirubin', 'tot_proteins', 'albumin', 'ag_ratio', 'sgpt', 'sgot', 'alkphos']])
```

```python
X = data.drop(['is_patient'],axis=1)
Y = data['is_patient']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, Ridge, ElasticNet



lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, Y_train)

ridge_model = Ridge(alpha=0.1)
ridge_model.fit(X_train, Y_train)


elasticnet_model = ElasticNet(alpha=0.1, l1_ratio=0.5)
```

```python
# Define L1 Regularization model
dis_model_l1 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7, activation='relu', input_shape=(10,)),
    tf.keras.layers.Dense(4, activation='relu', kernel_regularizer=tf.keras.regularizers.l1(0.01)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Define L2 Regularization model
dis_model_l2 = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7, activation='relu', input_shape=(10,)),
    tf.keras.layers.Dense(4, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.01)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Define Elastic Net Regularization model
dis_model_elastic = tf.keras.models.Sequential([
    tf.keras.layers.Dense(7, activation='relu', input_shape=(10,)),
    tf.keras.layers.Dense(4, activation='relu', kernel_regularizer=tf.keras.regularizers.l1_l2(l1=0.01, l2=0.5)),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
dis_model_l1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
dis_model_l2.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
dis_model_elastic.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
# L1 regularization model training
trained_dis_model_l1 = dis_model_l1.fit(X_train,Y_train,epochs=20,validation_data=(X_test,Y_test))
```

```python
# L2 Regularization model training
trained_dis_model_l2 = dis_model_l2.fit(X_train,Y_train,epochs=20,validation_data=(X_test,Y_test))
```
